```python
#getting the frequency of result column
ipl['result'].value_counts()
```

```python
#finding out the number of toss winner w.r.t to each team
ipl['toss_winner'].value_counts()
```

```python
#extracting the records where a team won batting first
batting_first=ipl[ipl['win_by_runs']!=0]
```

```python
batting_first.head()
```

```python
#making histogram(numerical value)
plt.figure(figsize=(5,5))
plt.hist(batting_first['win_by_runs'])
plt.title('distribution of runs')
plt.xlabel('Runs')
plt.show()
```

```python
batting_first['winner'].value_counts()[0:6]
```

```python
plt.figure(figsize=(9,6))
plt.bar(list(batting_first['winner'].value_counts()[0:4].keys()),list(ipl['winner'].value_counts()[0:4]), color=['green', 'blue','red','grey'])
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.pie(list(batting_first['winner'].value_counts()),labels=list(batting_first['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```

```python
#extracting those records where a team has won after batting second
batting_second=ipl[ipl['win_by_wickets']!=0]
```

```python
batting_second.head()
```

```python
plt.figure(figsize=(6,6))
plt.hist(batting_second['win_by_wickets'],bins=30)
plt.show()
```

```python
batting_second['winner'].value_counts()
```

```python
batting_second['winner'].value_counts()[0:3].keys()
```

```python
plt.figure(figsize=(6,6))
plt.pie(list(batting_second['winner'].value_counts()),labels=list(batting_first['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```
